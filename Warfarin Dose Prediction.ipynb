### Applying Label Encoding on Categorical Columns

```python

X_train_cat_le = pd.DataFrame(index=X_train_cat.index)

X_train_cat_le.head()
```

```python
X_train_cat.head()
```

```python
X_train_cat['Gender'].unique()
```

```python
Gender_encoder = {'female' : 1, 'male' : 2, 'Other' : 3}

X_train_cat_le['Gender'] = X_train_cat['Gender'].apply(lambda x : Gender_encoder[x])

X_train_cat_le.head()
```

```python
X_train_cat['Race (Reported)'].unique()
```

```python
Race_encoder = {'White' : 1, 'Caucasian' : 2, 'Japanese' : 3,'Han Chinese' : 4, 'Korean' : 5, 'Indian' : 6,
                'Chinese' : 7, 'Black other' : 8, 'Black or African American' : 9,'Black' : 10, 'Malay' : 11, 'Intermediate' : 12,
               'Other Mixed Race' : 13, 'Asian' : 14, 'Hispanic' : 15,'Black Caribbean' : 16, 'African-American' : 17, 'Other' : 18,
               'other' : 19, 'Black African' : 20, 'Other (Hungarian)' : 21,'Other (Black British)' : 22}

X_train_cat_le['Race (Reported)'] = X_train_cat['Race (Reported)'].apply(lambda x : Race_encoder[x])

X_train_cat_le.head()
```

```python
X_train_cat['Age'].unique()
```

```python
Age_encoder = {'70 - 79' : 1, '80 - 89' : 2, '50 - 59' : 3,'40 - 49' : 4, '60 - 69' : 5, '20 - 29' : 6,
                '30 - 39' : 7, '90+' : 8, '10 - 19' : 9
              }

X_train_cat_le['Age'] = X_train_cat['Age'].apply(lambda x : Age_encoder[x])

X_train_cat_le.head()
```

```python
X_train_cat['Cyp2C9 genotypes'].unique()
```

```python
Cyp2C9_encoder = {'*1/*1' : 1, '*2/*3' : 2, '*1/*2' : 3,'*1/*3' : 4, '*3/*3' : 5, '*2/*2' : 6,
                '*1/*6' : 7, '*1/*5' : 8, '*1/*11' : 9, '*1/*14':10, '*1/*13':11
              }

X_train_cat_le['Cyp2C9 genotypes'] = X_train_cat['Cyp2C9 genotypes'].apply(lambda x : Cyp2C9_encoder [x])

X_train_cat_le.head()
```

```python
X_train_cat['VKORC1 genotype: -1639 G>A (3673); chr16:31015190; rs9923231; C/T'].unique()
```

```python
VKORC1_encoder = {'G/G' : 1, 'A/A' : 2, 'nan' : 3,'A/G' : 4
              }
X_train_cat_le['VKORC1 genotype: -1639 G>A (3673); chr16:31015190; rs9923231; C/T'] = X_train_cat['VKORC1 genotype: -1639 G>A (3673); chr16:31015190; rs9923231; C/T'].apply(lambda x : VKORC1_encoder  [x])

X_train_cat_le.head()
```
